##### Feature Engineering

```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df = pd.read_csv("Travel.csv")
df.head()
```

```python
#checking for the missing values
df.isnull().sum()
```

```python
df['Gender'].value_counts()
```

```python
df["Gender"] = df["Gender"].replace("Fe Male","Female")
df['Gender'].value_counts()
```

```python
df["MaritalStatus"] = df['MaritalStatus'].replace("Unmarried","Single")
df['MaritalStatus'].value_counts()
```

```python
df['Total_visiting'] = df['NumberOfChildrenVisiting'] + df['NumberOfPersonVisiting']
df.drop(columns=['NumberOfChildrenVisiting','NumberOfPersonVisiting'],inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#handling the missing values
df['Age'].fillna(df['Age'].median(),inplace = True)
df['TypeofContact'].fillna(df['TypeofContact'].mode()[0],inplace = True)
df['DurationOfPitch'].fillna(df['DurationOfPitch'].median(),inplace = True)
df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].median(),inplace = True)
df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].median(),inplace = True)
df['NumberOfTrips'].fillna(df['NumberOfTrips'].median(),inplace = True)
df['Total_visiting'].fillna(df['Total_visiting'].median(),inplace = True)
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(),inplace = True)
```

```python
df.drop(df[['CustomerID']],axis = 1,inplace = True)
```

```python
df
```

##### Feature Extraction

```python
#Extract the categorical features
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
print(f"The no of categorical features are: {len(categorical_features)}")
```

```python
#extract Numerical features
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']
print(f"The no of numerical features are:{len(numerical_features)}")
```

```python
#Independent and dependent features
X = df.drop(df[['ProdTaken']],axis = 1)
y = df['ProdTaken']
```

```python
#Implement train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
#Encoding the categorical and standardizing the numerical features
numerical_features = X.select_dtypes(exclude = "object").columns
categorical_features = X.select_dtypes(include = "object").columns

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

ohe = OneHotEncoder(drop = 'first')
scaler = StandardScaler()

columnTrans = ColumnTransformer([
    ("OnehotEncoder",ohe,categorical_features),
    ("StandardScaler",scaler,numerical_features)
])
```

```python
X_train = columnTrans.fit_transform(X_train)
```

```python
X_test = columnTrans.transform(X_test)
```

##### Model training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

models = {
    "DecisionTree":DecisionTreeClassifier(),
    "RandomForest":RandomForestClassifier()
}

for i in range(len(list(models))):
    model =list(models.values())[i]
    model.fit(X_train,y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    y_train_acc = accuracy_score(y_train,y_train_pred)
    y_test_acc = accuracy_score(y_test,y_test_pred)
    print(f"\nThe model is {list(models.keys())[i]}")
    print(f"The Training accuracy is {y_train_acc}")
    print(f"The Testing accuracy is {y_test_acc}")
```
